```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./sales_data_sample.csv', encoding='unicode_escape')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
X=df[['QUANTITYORDERED','PRICEEACH','SALES']]
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x_scaled=sc.fit_transform(X)
x_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
#inertia stores all the values of SumofSquaredError
inertia=[]
K=range(2,11)  # start at 2 clusters
for k in K:
    kmeans=KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(x_scaled)
    inertia.append(kmeans.inertia_)
```

```python
#plotting 
plt.figure(figsize=(12,8))
plt.style.use('ggplot')
plt.plot(K,inertia,'bo-')
plt.xlabel('K')
plt.ylabel('Inertia (SSE)')
plt.title('K-Means Elbow Plot')
plt.show()
```

```python
#optimal k value=4
optimal_k=4
kmeans=KMeans(n_clusters=optimal_k, n_init=10, random_state=42)
df['Clusters']=kmeans.fit_predict(x_scaled)
df[['QUANTITYORDERED','PRICEEACH','SALES','Clusters']]
```
